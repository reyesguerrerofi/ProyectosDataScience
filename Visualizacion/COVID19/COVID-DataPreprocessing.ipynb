### Canada muestra una discrepancia en sus datos. En confirmados y muertes sus registros son por provincia, en cambio con recuperados es general. Hay que arreglarlo

```python
#Obtenemos el total de confirmados y muertos por fecha
confirmados_canada = confirmados[confirmados["Country/Region"]=="Canada"].groupby("fecha").sum()[["confirmado"]]
muertos_canada = muertes[muertes["Country/Region"]=="Canada"].groupby("fecha").sum()[["muertes"]]
```

```python
#Se extraen las columnas de las respectivas tablas
canada_recuperados = recuperados[recuperados["Country/Region"]=="Canada"]  
canada_template = canada_recuperados[canada_recuperados.columns[:-1]].reset_index(drop=True)
```

```python
# Union de los datos de confirmados y muertos
confirmados_canada  = canada_template.merge(confirmados_canada, how="inner", left_on="fecha", right_index=True)
muertos_canada = canada_template.merge(muertos_canada,how="inner",left_on="fecha",right_index=True)
```

```python
#Agregar los datos a confirmados y muertos
confirmados = confirmados[confirmados["Country/Region"] != "Canada"].append(confirmados_canada)
muertes = muertes[muertes["Country/Region"] != "Canada"].append(muertos_canada)
```

Todo lo anterior fue para que confirmados y muertes ya no esten catalogados por provincia y asi poder hacer el merge

```python
data = confirmados.merge(muertes,how="inner",on=["Country/Region","Province/State","fecha"]).merge(recuperados,how="inner",on=["Country/Region","Province/State","fecha"])
data["recuperados"] = data["recuperados"].astype("int")
data = data[["Province/State","Country/Region","fecha","Lat","Long","confirmado","muertes","recuperados"]]
```
