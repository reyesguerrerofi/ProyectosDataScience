# Creacion de nubes de palabras.

```python
import wordcloud
from wordcloud import WordCloud, STOPWORDS
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')

#Crear lista de palabras

stopwords = set(STOPWORDS)
stopwords.update(["br","href"])
textt = " ".join(review for review in df.Text)
wordcloud = WordCloud(stopwords=stopwords).generate(textt)

plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.savefig("wordcloud1.png")
plt.show()
```

Las palabras que se pueden ver son "Taste"(Probar), "Product" (Producto), "Amazon", "Love"(Amar), "Flavor"(Sabor).
Debido a que la proporcion de comentarios positivos es mayor parece ser que las palabras mas comunes tambien se atribuyen a comentarios positivos.

## Clasificacion de los reviews.
Para facilidad de la clasificacion en el modelo se estableceran los parametros **positivo** y **negativo**.
* Los votos positivos seran +1 y negativos -1
* Los positivos seran los que tengan la columna Score > 3.
* Los negativos Score < 3.
* Los neutrales Score = 3 se les hara drop. Ya que el modelo solo clasificara votos negativos y positivos.

```python
#Votando los Score ==3

df = df[df["Score"] != 3]

#Se aplica el sistema de puntajes.

df["sentimiento"] = df["Score"].apply(lambda rating: +1 if rating > 3 else -1)
```

```python
df.head()
```

## Mas analisis

```python
#Separamos positivo y negativo

positivo = df[df["sentimiento"]==1]
negativo = df[df["sentimiento"]==-1]
negativo = negativo[negativo["Summary"].notna()]
```

```python
#Ahora si se hara un analisis de sentimientos por separado

#Positivo

stopwords = set(STOPWORDS)
stopwords.update(["br","href","good","great"])#Son palabras que se pueden omitir y no altera el significado de
#la frase. good y great se removieron porque estas fueron incluidas en los sentimientos negativos.

positivo = " ".join(review for review in positivo.Summary)
wordcloud2 = WordCloud(stopwords=stopwords).generate(positivo)

plt.imshow(wordcloud2, interpolation="bilinear")
plt.axis("off")
plt.savefig("wordcloudPositivo.png")
plt.show()
```

Palabras como "love" (Amar), "Delicious"(Delicioso),"taste"(Probar),"best"(El mejor),"product"(Producto),
"Awesome"(Asombroso) estan relacionados a sentimientos positivos.

```python
#Negativos


negativo = " ".join(review for review in negativo.Summary)
wordcloud3 = WordCloud(stopwords=stopwords).generate(negativo)

plt.imshow(wordcloud3, interpolation="bilinear")
plt.axis("off")
plt.savefig("wordcloudNegativo.png")
plt.show()
```

Palabras como "dissapointed"(Decepcionadx),"Yuck"(Asco) son palabras asociadas a negativo.

Retomando un poco lo de "good" y "great" puede ser usado como "not good" por lo que no se utiliza en positivo ni negativo.

```python
#Distribucion de negativo y positivo

df["sentimiento"] = df["sentimiento"].replace({-1:"negativo"})
df["sentimiento"] = df["sentimiento"].replace({1:"positivo"})

sns.histplot(df["sentimiento"])
```
